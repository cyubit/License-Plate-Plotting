```python
import numpy as np
from matplotlib import pyplot as plt
import warnings
warnings.simplefilter('ignore', np.RankWarning)
```

```python
def getInput():
    plates = []
    while input("Add a plate? (y/n) ") == "y":
        plates.append(list(map(int,input("Enter the plate #: "))))
    return plates

def main():
    plates = getInput()
    figure, axis = plt.subplots(1,3,figsize=(15, 5))
    axis[0].set_title("Piecewise Function")
    axis[1].set_title("Scatter Plot")
    axis[2].set_title("Extended Piecewise")
    for plate in plates: 
        axis[0].plot(range(1, len(plate) + 1), plate, marker='o', label=''.join(map(str,plate)))
        axis[1].scatter(range(1, len(plate) + 1), plate, marker='o', label=''.join(map(str,plate)))
        x = [*range(1, len(plate) + 1)]
        for p in range(len(plate)):
            slope, intercept = np.polyfit(x[p:p+2], plate[p:p+2], 1)
            abline_values = [slope * i + intercept for i in x]
            axis[2].plot(x, abline_values)
        axis[2].scatter(range(1, len(plate) + 1), plate, marker='o', label=''.join(map(str,plate)))

            # axis[2].plot(plate[x],plate[x+1], marker='o', label=''.join(map(str,plate)))
        # plt.plot(range(1, len(plate) + 1), plate, marker='o', label=''.join(map(str,plate)))
        # plt.scatter(range(1, len(plate) + 1), plate, marker='o', label=''.join(map(str,plate)))
    for i in range(3):
        axis[i].grid()
    plt.tight_layout()
    plt.xlabel("Number Position")
    plt.ylabel("Plate Number")
    plt.xlim(0,7)
    plt.ylim(0,10)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.show()

main()
```
